```python
corrMatt = df[["Order_ID","Order_Priority","Order_Quantity","Sales","Ship_Mode","Region","Customer_Segment","Product_Category","Product_Sub-Category","Product_Name","Product_Container","Profit"]].corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt, mask=mask,vmax=.8, square=True,annot=True)
```

Significance of the parameters on each other, looks like only sales and order quantity are related to profits and have significance over it. We have seen a linear plot between the sales and the profits suggesting a linear relationsip between them but the slope of the line is not satisfying since it is very low.
